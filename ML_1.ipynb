**1. Гистограмма для числовой переменной Age:**

```python
sns.histplot(x=df['age'], kde=True)
plt.title('Распределение age')
plt.xlabel('age')
plt.ylabel('Частота')
plt.show()
```

**2. Столбчатая диаграмма для категориальной переменной race:**

```python
sns.countplot(x='race', data=df)
plt.title('Распределение race')
plt.xlabel('Распределение race')
plt.ylabel('Количество')
plt.show()
```

**3. Боксплот для числовой переменной education-num:**

```python
sns.boxplot(x=df['education-num'])
plt.title('Boxplot education-num')
plt.xlabel('education-num')
plt.show()
```

**Ваши наблюдения:**
1. Распределение возраста сосредоточено не как нормальное распределение. Средний возраст составляет около 38 лет.

2. Распределение рас говорит о том, что по статистике в большинство принимающих участиек в таблице составляют белые, а потом с сильным разрывом чёрные, азиаты и остальные.

3. Боксплот указывает о неоднородном распределении уровня образования, среди учтённых в статистике.

Корреляционная матрица

```python
correlation_matrix = df[['age', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week', 'income >50K']].corr()
```

Визуализация тепловой карты

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()
```

По матрице видно, что сильная положительная корреляция существует между уровнем оюразования и размером заработка. Чуть менее она выражена между возрастом и доходом, между часами рабочей недели и размером дохода. В остальных случаях корреляция слабо выражена.
